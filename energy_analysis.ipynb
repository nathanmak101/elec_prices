```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def set_date_index(df: pd.DataFrame, date_column_name: str, date_format_str: str) -> pd.DataFrame:
    df = df.rename(columns={date_column_name: "date"})
    df = df.set_index("date")
    df.index = pd.to_datetime(df.index, format = date_format_str)
    return df
```

```python
pd.ExcelFile("state_energy_data_wide.xlsx").sheet_names
```

```python
all_sheets = pd.read_excel("state_energy_data_wide.xlsx", sheet_name=None)

all_sectors = set_date_index(all_sheets["all_sectors"], "date", "%Y-%m-%d")
commercial = set_date_index(all_sheets["commercial"], "date", "%Y-%m-%d")
industrial = set_date_index(all_sheets["industrial"], "date", "%Y-%m-%d")
other = set_date_index(all_sheets["other"], "date", "%Y-%m-%d")
residential = set_date_index(all_sheets["residential"], "date", "%Y-%m-%d")
transportation = set_date_index(all_sheets["transportation"], "date", "%Y-%m-%d")
```

```python
df_list = [all_sectors, commercial, industrial, other, residential, transportation]
df_list_names = ['all_sectors', 'commercial', 'industrial', 'other', 'residential', 'transportation']

for data, name in zip(df_list, df_list_names):
    plt.figure(figsize=(10, 6))
    plt.plot(data)
    plt.title(name)
    plt.show()
```
